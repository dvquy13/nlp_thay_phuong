```python
import pandas as pd
```

```python
def load_sent(raw_txt):
    return list(filter(lambda x: x != '\n', raw_txt))

def split_word(s):
    return [x for x in s.split() if x != " "]
assert split_word("\tThe/at jury/nn  presentments/nns Georgia's/np$ ") == [
    "The/at",
    "jury/nn",
    "presentments/nns",
    "Georgia's/np$"
]

def get_tag(s):
    return s[s.rfind("/")+1:]
assert get_tag("The/at") == ("at")
assert get_tag("county/nn-t1") == ("nn-t1")
assert get_tag("georgia's/np$") == ("np$")
```

```python
with open("./data/brown_dataset/tagged_train.txt") as f:
    raw = f.readlines()
raw = load_sent(raw)
```

```python
raw[:5]
```

```python
tmp = raw
foo = [map(get_tag, x) for x in map(split_word, tmp)]
```

```python
foo
```

```python
from collections import Counter
counter = Counter()
for sent in foo:
    for word in sent:
        counter[word] += 1
```

```python
count_df = pd.DataFrame.from_dict(dict(counter), orient='index').reset_index().rename(columns={0: "count", "index": "tag"})
count_df.sort_values("count", ascending=False)
```

```python
def distill_tag(s):
    s = s.replace("fw-", "")
    hyphen_pos = s.find("-")
    if hyphen_pos != -1:
        rm_hyphen = s[:hyphen_pos]
    else:
        rm_hyphen = s
    return rm_hyphen.split("+")
assert distill_tag("nn+md") == ["nn", "md"]
```

```python
test = count_df.copy()
test['tag'] = test['tag'].apply(distill_tag)
```

```python
import string
puncs = set(string.punctuation) - set(["$"])
```

```python
unstack = test.set_index('count')['tag'].apply(pd.Series).stack().reset_index(level=-1, drop=True).reset_index().rename(columns={0: "tag"})
tag_count = unstack.groupby(['tag'])['count'].sum().sort_values(ascending=False).to_frame().reset_index()
```

```python
tag_count['tag'] = tag_count['tag'].apply(lambda x: "".join(chr for chr in x if not chr in puncs))
tag_count = tag_count[(tag_count['tag'] != "")]
count_total = tag_count['count'].sum()
tag_count['portion'] = tag_count['count'] / count_total * 100
tag_count
```

```python
tag_count = tag_count.groupby(['tag']).sum().sort_values('count', ascending=False).reset_index()
```
